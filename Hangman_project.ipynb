```python
word_list=['vishal', 'hardik']


import random
word=random.choice(word_list)

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

blanks=[]
for letter in word:
    blanks.append('_')
print(blanks)

lives=6

end_of_game=False

while not end_of_game:
    var=input('Guess a letter: ').lower()
    
    
    

    for position in range(len(word)):
        letter=word[position]
        if letter==var:
            blanks[position]=letter
            print(blanks)
    
    if var not in word:
        lives-=1
        if lives==0:
            end_of_game=True

            

    
    
    if '_' not in blanks:
        end_of_game=True
        
        
    print(stages[lives])
        
        
```
